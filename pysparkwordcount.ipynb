```python
import pyspark 
from pyspark import SparkContext
import os
import sys
os.environ['PYSPARK_DRIVER_PYTHON_OPTS']='notebook'
os.environ['PYSPARK_DRIVER_PYTHON']=sys.executable
os.environ['PYSPARK_PYTHON']=sys.executable
```

```python
sc= SparkContext()
```

```python
text_file = sc.textFile(r"C:\Users\91976\Desktop\big_data_pro\sample_txt.txt")
text_file.collect()
```

```python
newp = text_file.flatMap(lambda line: line.split(","))
```

```python
newp.collect()
```

```python
counts = text_file.flatMap(lambda line: line.split(" "))
counts.collect()
```

```python
count1 = counts.map(lambda word: (word, 1))
count1.collect()
```

```python
count3 = count1.reduceByKey(lambda x, y: x + y)
count3.collect()
```

```python
count3.filter(lambda x: x[0] not in ["for", "and", "not", "on", "the", "as", "of", "is"]).collect()
```
